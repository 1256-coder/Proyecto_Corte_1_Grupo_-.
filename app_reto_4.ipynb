```python
# PROGRAMA DE LEALTAD - VERSIÓN PRINCIPIANTE
# Tienda de ropa

print("=" * 40)
print("PROGRAMA DE LEALTAD")
print("Registro de 5 clientes")
print("=" * 40)

# Listas para guardar la información
nombres = []
compras = []
categorias = []
bonos = []

# Registrar 5 clientes
for i in range(5):
    print(f"\n--- Cliente {i+1} ---")
    
    # Pedir nombre
    nombre = input("Nombre del cliente: ")
    nombres.append(nombre)
    
    # Pedir valor de compra
    valor = float(input("Valor total de compra: $"))
    compras.append(valor)
    
    # Asignar categoría
    if valor < 500000:
        categoria = "Bronce"
        bono = 0
    else:
        categoria = "Oro"
        bono = valor * 0.10  # 10% de bono
    
    categorias.append(categoria)
    bonos.append(bono)
    
    # Mostrar resultado
    print(f"Categoría: {categoria}")
    if bono > 0:
        print(f"Bono de regalo: ${bono}")

# ORDENAR ALFABÉTICAMENTE (método burbuja simple)
print("\n" + "=" * 40)
print("ORDENANDO CLIENTES...")
print("=" * 40)

# Ordenar nombres junto con sus datos
for i in range(4):  # 4 porque tenemos 5 elementos (0-4)
    for j in range(4 - i):
        if nombres[j] > nombres[j + 1]:  # Comparar nombres
            # Intercambiar nombres
            nombres[j], nombres[j + 1] = nombres[j + 1], nombres[j]
            # Intercambiar compras
            compras[j], compras[j + 1] = compras[j + 1], compras[j]
            # Intercambiar categorias
            categorias[j], categorias[j + 1] = categorias[j + 1], categorias[j]
            # Intercambiar bonos
            bonos[j], bonos[j + 1] = bonos[j + 1], bonos[j]

# MOSTRAR REPORTE FINAL
print("\n" + "=" * 50)
print("REPORTE DE CLIENTES (orden alfabético)")
print("=" * 50)

print("No. | Cliente           | Compra     | Categoría | Bono")
print("-" * 50)

for i in range(5):
    # Formatear para que se vea ordenado
    nombre_mostrar = nombres[i][:15] + "..." if len(nombres[i]) > 15 else nombres[i]
    print(f"{i+1:2d}  | {nombre_mostrar:15s} | ${compras[i]:8.0f} | {categorias[i]:7s} | ${bonos[i]:8.0f}")

# Calcular totales
total_ventas = 0
total_bonos = 0
clientes_oro = 0
clientes_bronce = 0

for i in range(5):
    total_ventas = total_ventas + compras[i]
    total_bonos = total_bonos + bonos[i]
    if categorias[i] == "Oro":
        clientes_oro = clientes_oro + 1
    else:
        clientes_bronce = clientes_bronce + 1

# Mostrar resumen
print("\n" + "=" * 40)
print("RESUMEN")
print("=" * 40)
print(f"Total ventas: ${total_ventas}")
print(f"Total bonos: ${total_bonos}")
print(f"Clientes Oro: {clientes_oro}")
print(f"Clientes Bronce: {clientes_bronce}")

print("\n¡Programa finalizado!")
```
