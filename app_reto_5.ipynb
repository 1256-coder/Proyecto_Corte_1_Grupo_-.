```python
# NÓMINA-EXPRESS 
# Liquidación de nómina semanal para PyME

print("=" * 50)
print("NÓMINA-EXPRESS")
print("Liquidación de nómina semanal")
print("=" * 50)

# Constantes
VALOR_HORA = 20000  # $20,000 por hora (valor simulado)
SMMLV = 1300000     # Salario Mínimo Mensual simulado
AUXILIO_TRANSPORTE = 162000  # Auxilio de transporte simulado

# Listas para guardar información
nombres = []
horas_trabajadas = []
horas_normales = []
horas_extras = []
sueldo_base = []
sueldo_total = []
auxilio_aplicado = []

# Ingresar datos de 5 empleados
for i in range(5):
    print(f"\n--- Empleado {i+1} ---")
    
    # Pedir nombre
    nombre = input("Nombre del empleado: ")
    nombres.append(nombre)
    
    # Pedir horas trabajadas
    horas = float(input("Horas trabajadas en la semana: "))
    horas_trabajadas.append(horas)
    
    # Calcular horas normales y extras
    if horas <= 48:
        normales = horas
        extras = 0
    else:
        normales = 48
        extras = horas - 48
    
    horas_normales.append(normales)
    horas_extras.append(extras)
    
    # Calcular sueldo base
    valor_normales = normales * VALOR_HORA
    valor_extras = extras * (VALOR_HORA * 1.25)  # 25% más
    sueldo_base_calc = valor_normales + valor_extras
    
    sueldo_base.append(sueldo_base_calc)
    
    # Verificar si aplica auxilio de transporte (menos de 2 SMMLV)
    if sueldo_base_calc < (2 * SMMLV):
        sueldo_final = sueldo_base_calc + AUXILIO_TRANSPORTE
        auxilio = "Sí"
    else:
        sueldo_final = sueldo_base_calc
        auxilio = "No"
    
    sueldo_total.append(sueldo_final)
    auxilio_aplicado.append(auxilio)
    
    # Mostrar resumen del empleado
    print(f"\n  Resumen para {nombre}:")
    print(f"  Horas normales: {normales:.0f}")
    print(f"  Horas extras: {extras:.0f}")
    print(f"  Sueldo base: ${sueldo_base_calc:,.0f}")
    print(f"  Auxilio transporte: {auxilio}")
    print(f"  Sueldo total: ${sueldo_final:,.0f}")

# MOSTRAR DESPRENDIBLE TOTAL DE LA EMPRESA
print("\n" + "=" * 70)
print("DESPRENDIBLE DE PAGO - NÓMINA SEMANAL")
print("=" * 70)

# Encabezado de la tabla
print(f"{'No.':<4} {'EMPLEADO':<15} {'HORAS':<6} {'NORM.':<6} {'EXTRAS':<6} {'BASE':<12} {'AUXILIO':<8} {'TOTAL':<12}")
print("-" * 70)

# Mostrar cada empleado
for i in range(5):
    print(f"{i+1:<4} {nombres[i][:14]:<15} {horas_trabajadas[i]:<6.0f} {horas_normales[i]:<6.0f} {horas_extras[i]:<6.0f} ${sueldo_base[i]:<11,.0f} {auxilio_aplicado[i]:<8} ${sueldo_total[i]:<11,.0f}")

# Calcular totales
print("-" * 70)

total_horas = 0
total_base = 0
total_pagar = 0
total_auxilios = 0
total_extras = 0

for i in range(5):
    total_horas = total_horas + horas_trabajadas[i]
    total_base = total_base + sueldo_base[i]
    total_pagar = total_pagar + sueldo_total[i]
    total_extras = total_extras + horas_extras[i]
    if auxilio_aplicado[i] == "Sí":
        total_auxilios = total_auxilios + AUXILIO_TRANSPORTE

# Mostrar totales
print(f"{'TOTALES:':<19} {total_horas:<6.0f} {'':<6} {total_extras:<6.0f} ${total_base:<11,.0f} {'':<8} ${total_pagar:<11,.0f}")
print("=" * 70)

# Resumen adicional
print("\n" + "=" * 70)
print("RESUMEN DE LA NÓMINA")
print("=" * 70)
print(f"Total empleados: 5")
print(f"Total horas trabajadas: {total_horas:.0f} horas")
print(f"Total horas extras: {total_extras:.0f} horas")
print(f"Total sueldos base: ${total_base:,.0f}")
print(f"Total auxilio de transporte: ${total_auxilios:,.0f}")
print(f"TOTAL A PAGAR: ${total_pagar:,.0f}")
print("=" * 70)

print("\n¡Liquidación de nómina completada!")
```
